```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, Flatten, BatchNormalization, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
import itertools
import os
import shutil
import random
import glob
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
%matplotlib inline
```

```python
#Organize data into train, valid, test dirs
os.chdir('/home/kubera/Documents/DLProjects/dogs-vs-cats/data')
if os.path.isdir('train/dog') is False:
  os.makedirs('train/dog')
  os.makedirs('train/cat')
  os.makedirs('valid/dog')
  os.makedirs('valid/cat')
  os.makedirs('test/dog')
  os.makedirs('test/cat')
```

```python
# List all cat and dog files
cat_files = glob.glob('cat*')
dog_files = glob.glob('dog*')

# Debugging: Print initial counts
print("Initial cat count:", len(cat_files))
print("Initial dog count:", len(dog_files))

# Shuffle the lists
random.shuffle(cat_files)
random.shuffle(dog_files)

# Debugging: Print initial counts
print("Initial cat count:", len(cat_files))
print("Initial dog count:", len(dog_files))

# Define the sizes for each set
train_size = 500
valid_size = 100
test_size = 50

# Move files to train
for c in cat_files[:train_size]:
    shutil.move(c, 'train/cat')
for d in dog_files[:train_size]:
    shutil.move(d, 'train/dog')

# Debugging: Print initial counts
print("Initial cat count:", len(cat_files))
print("Initial dog count:", len(dog_files))

# Move files to validation
for c in cat_files[train_size:train_size + valid_size]:
    shutil.move(c, 'valid/cat')
for d in dog_files[train_size:train_size + valid_size]:
    shutil.move(d, 'valid/dog')

# Debugging: Print initial counts
print("Initial cat count:", len(cat_files))
print("Initial dog count:", len(dog_files))

# Move files to test
for c in cat_files[train_size + valid_size:train_size + valid_size + test_size]:
    shutil.move(c, 'test/cat')
for d in dog_files[train_size + valid_size:train_size + valid_size + test_size]:
    shutil.move(d, 'test/dog')

# Debugging: Print initial counts
print("Initial cat count:", len(cat_files))
print("Initial dog count:", len(dog_files))
```

```python
train_path = '/home/kubera/Documents/DLProjects/dogs-vs-cats/data/train'
valid_path = '/home/kubera/Documents/DLProjects/dogs-vs-cats/data/valid'
test_path = '/home/kubera/Documents/DLProjects/dogs-vs-cats/data/test'
```

```python
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=train_path, target_size=(224,224), classes=['cat','dog'], batch_size=10)

valid_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=valid_path, target_size=(224,224), classes=['cat','dog'], batch_size=10)

test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=test_path, target_size=(224,224), classes=['cat','dog'], batch_size=10, shuffle=False)
```

```python
assert train_batches.n == 1000
assert valid_batches.n == 200
assert test_batches.n == 100
assert train_batches.num_classes == valid_batches.num_classes == test_batches.num_classes == 2
```

```python
imgs, labels = next(train_batches)
```

```python
#This function will plot images in the form of a grid with 1 row and 10 columns where images are places them using matplotliv

def plotImages(images_arr): 
    fig, axes = plt.subplots(1, 10, figsize=(20, 20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(imgs)
print(labels)
```

```python
#Build and train a CNN
model = Sequential([
    Conv2D(filters=32, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(224,224,3)), 
    MaxPool2D(pool_size=(2,2), strides=2),
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same'),
    MaxPool2D(pool_size=(2,2), strides=2),
    Flatten(),
    Dense(units=2, activation='softmax'),
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=train_batches, steps_per_epoch=len(train_batches), validation_data = valid_batches, validation_steps=len(valid_batches), epochs=10, verbose=2)
```

```python
#Predict 
test_imgs, test_labels = next(test_batches)
plotImages(test_imgs)
print(test_labels)
```

```python
def plot_confusion_matrix(cm, classes,normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
test_batches.class_indices
```

```python
cm_plot_labels = ['cat', 'dog']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```
